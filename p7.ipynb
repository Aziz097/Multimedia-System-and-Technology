```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpy_img_ch3 = np.array([[[255, 0, 0], [0, 255, 0], [0, 0, 255]],
                            [[255, 255, 0], [0, 255, 255], [255, 0, 255]],
                            [[192, 192, 192], [128, 128, 128], [0, 0, 0]]]
                            , dtype=np.uint8)



plt.imshow(numpy_img_ch3)
plt.axis('off')
plt.show()
```

```python
# change 3 channels to 1 channel using cv2 mean and axis
numpy_img_ch1 = np.mean(numpy_img_ch3, axis=1, dtype=np.uint8)
print(numpy_img_ch1)
plt.imshow(numpy_img_ch1, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import os
import cv2

# load image dengan os get cwd
img_path = os.path.join(os.getcwd(), 'download.jpeg')

#check if file exists
if_exists = os.path.isfile(img_path)
print(if_exists)
```

```python
#load image cv2
img = cv2.imread(img_path)

#print shape
print(img.shape) 
#plot figure nya
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
#gunakan numpy untuk merubah pixel di tengah menjadi warna merah
img_height, img_width, _ = img.shape
center_y, center_x = img_height // 2, img_width // 2
img[center_y-40:center_y+50, center_x-50:center_x+50] = [0, 0, 255] # BGR format
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Image with Red Square in Center')
plt.axis('off')
plt.show()
```

```python
#gunakan numpy untuk merubah pixel di pojok kiri atas menjadi warna hijau
img[0:100, 0:100] = [0, 255, 0] # BGR format
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Image with Green Square in Top-Left Corner')
plt.axis('off')
plt.show()
```

```python
#gunakan numpy untuk merubah pixel di pojok kanan atas menjadi warna biru
img[0:100, img_width-100:img_width] = [255, 0, 0] # BGR format
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Image with Blue Square in Top-Right Corner')
plt.axis('off')
plt.show()
```

```python
border_black = 10
border_red = 5
img_with_border = cv2.copyMakeBorder(img, border_black, border_black, border_black, border_black, cv2.BORDER_CONSTANT, value=[0, 0, 0])
img_with_border = cv2.copyMakeBorder(img_with_border, border_red, border_red, border_red, border_red, cv2.BORDER_CONSTANT, value=[0, 0, 255])
plt.imshow(cv2.cvtColor(img_with_border, cv2.COLOR_BGR2RGB))
plt.title('Image with Borders')
plt.axis('off')
plt.show()
```

```python
print(img_with_border.shape) 
```

```python
# crop image di tengah 450x450
crop_size = 350
start_y = (img.shape[0] - crop_size) // 2
start_x = (img.shape[1] - crop_size) // 2
cropped_img = img[start_y:start_y+crop_size, start_x:start_x+crop_size]
plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')
plt.axis('off')
plt.show()
```

```python
import os
import cv2

# load image dengan os get cwd
path = os.path.join(os.getcwd(), 'download.jpeg')

#check if file exists
if_exists = os.path.isfile(img_path)
print(if_exists)

#load image cv2
image = cv2.imread(path)

#print shape
print(img.shape) 
#plot figure nya
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
#naikan nilai pixel r nya 15 using np clip
image_brighter = image.copy()
image_brighter[:, :, 2] = np.clip(image_brighter[:, :, 2] + 15, 0, 255)
image_brighter[:, :, 1] = np.clip(image_brighter[:, :, 1] + 15, 0, 255)
image_brighter[:, :, 0] = np.clip(image_brighter[:, :, 0] + 15, 0, 255)
plt.imshow(cv2.cvtColor(image_brighter, cv2.COLOR_BGR2RGB))
plt.title('Brighter Image')
plt.axis('off')
plt.show()
```

```python
crop_size = 7
start_y = (image_brighter.shape[0] - crop_size+1) // 2
start_x = (image_brighter.shape[1] - crop_size) // 2
cropped_image = image_brighter[start_y:start_y+crop_size, start_x:start_x+crop_size]
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')
plt.axis('off')
plt.show()
```
